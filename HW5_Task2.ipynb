### Consider “naïve” method: tomorrow is the same as today
### %change tomorrow = %change today


```python
new_df = df.copy()

#Pick one year: 2017 
new_df['Date'] = pd.to_datetime(new_df['Date'])
start_date = '01-01-2017'
end_date = '12-31-2017'
mask = (new_df['Date'] > start_date) & (new_df['Date'] <= end_date)
new_df = new_df.loc[mask]


new_df.rename(columns={'Adj Close': 'Price'}, inplace = True)
new_df = new_df[['Date', 'Price']]
new_df['pct_change'] = new_df['Price'].pct_change()
new_df['pct_change'].fillna(0, inplace = True)
new_df['predicted'] = new_df['pct_change'].shift(1)
new_df['predicted'].fillna(0, inplace = True)
new_df
```

```python
new_df['|actual- predicted|'] = abs(new_df['pct_change'] - new_df['predicted'])
new_df['actual- predicted'] = new_df['pct_change'] - new_df['predicted']
```
