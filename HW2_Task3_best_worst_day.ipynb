```python
def compute_percent_df(df):
    try:
        new_df = df.copy()
        new_df.rename(columns={'Adj Close': 'Price'}, inplace = True)
        new_df = new_df[['Date', 'Price']]
        new_df['price_change(%)'] = new_df['Price'].pct_change() * 100
        new_df.fillna(0, inplace = True)
        return new_df
    except Exception as e:
        print(e)
        return None 
```

### (a)	take your stock (Oct  2, 2013 – Aug 31, 2018)
### (b) compute daily percent and define good/bad days according to %change  

```python
df['Date'] = pd.to_datetime(df['Date']) 
from_ts = '2013-10-02'
to_ts = '2018-08-31'
 
mask = (df['Date'] >= from_ts) & (df['Date'] <= to_ts)
df = df.loc[mask]

new_df = compute_percent_df(df)
new_df['good/bad'] = np.where(new_df['price_change(%)'] >= 0,'good','bad')
new_df = new_df.reset_index()
new_df
```

### (c)	compute total accumulation if you started with 100 dollars on October 2, 2013 assuming that you know the performance on the next day

```python
def buy(cost,price):
    shares = cost / price
    return shares

def sell(cost,shares, price):
    revenue = shares * price
    profit = revenue - cost
    return profit

def profit_calculator(new_df):
    cost = 100    
    shares = []
    profit = []
    
    #new_df['buy/sell'] = np.where(new_df['price_change(%)'] >= 0,'buy','sell')
    conditions = [(new_df['good/bad']== 'good') & (new_df['good/bad'].shift(-1) == 'good'), (new_df['good/bad']== 'good') & (new_df['good/bad'].shift(-1) == 'bad'),(new_df['good/bad']== 'bad') & (new_df['good/bad'].shift(-1) == 'good'),(new_df['good/bad']== 'bad') & (new_df['good/bad'].shift(-1) == 'bad')]
    choices = ['hold','sell','buy','hold']
    new_df['buy/sell'] = np.select(conditions,choices)

    if new_df.loc[0]['buy/sell']== 'sell':
        new_df.at[0, 'buy/sell'] = 'hold'
        
    
    for row in range(len(new_df)):
        if new_df.loc[row]['buy/sell'] == 'buy':
            shares.append(buy(cost,new_df.loc[row]['Price']))    
        elif new_df.loc[row]['buy/sell'] == 'sell':
            shares.append(0)      
        else:
            if row ==0:
                shares.append(0)
            else:    
                shares.append(shares[len(shares)-1])
    new_df['Shares']= shares
    
    for row in range(len(new_df)):
        if new_df.loc[row]['buy/sell'] == 'sell':
            profit.append(sell(cost,new_df.loc[row-1]['Shares'],new_df.loc[row]['Price']))
        else:
            profit.append(0)
    
    new_df['Profit']= profit
    
    return new_df
```

```python
profit_df = profit_calculator(new_df)
```

```python
# count the total number of transactions
trans_df = new_df[profit_df['buy/sell']=='sell']
total_trans = len(trans_df)  
total_profit = sum(trans_df['Profit'])

print("trans = ", total_trans)
print("Profit = ", total_profit )
```
