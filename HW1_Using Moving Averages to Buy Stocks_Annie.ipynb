```python
# parsing date column
df['Year'] = pd.DatetimeIndex(df['Date']).year
df['Month'] = pd.DatetimeIndex(df['Date']).month
df['Day_name'] = pd.DatetimeIndex(df['Date']).weekday_name
df['Date'] = pd.DatetimeIndex(df['Date']).day
df.head(5) 
```

```python
# compute average close price by year
df_year_close = df[['Year', 'Close']]
avg_by_year = df_year_close.groupby('Year')['Close'].mean()
avg_by_year
```

```python
# compute average close price by year and month
df_year_month_close = df[['Year', 'Month', 'Close']]
avg_by_year_month = df_year_month_close.groupby(['Year', 'Month'])['Close'].mean()
avg_by_year_month
```

```python
# reset index
avg_by_year_month = avg_by_year_month.reset_index()
avg_by_year_month
```

```python
# rename column name(s)
avg_by_year_month.rename(columns = {'Close': 'mean_month'},inplace = True)
avg_by_year_month.head(5)
```

```python
# percent change
df['%_change'] = df['Adj Close'].pct_change()
# cleaning data: replace "NaN" with "0"
df.fillna(0, inplace = True)
df.head(5)
```

```python
# computing moving average (W=14 and 50)
# short term moving average (14 days):st_ma --> s_win = 14 (short window length)
# long term moving average (15 days) :lt_ma --> l_win = 50 (long window length)

s_win = 14 
l_win = 50

df['Adj Close'].rolling(window =14).mean()
df['st_ma'] = df['Adj Close'].rolling(window = s_win).mean()
df['lt_ma'] = df['Adj Close'].rolling(window = l_win).mean()

# remove the first 50 days' data, since each moving average need 50 days to compute, NaN cannot be used
df = df.tail(len(df)-50)
# buy and sell signals
df['Buy'] = df['st_ma'] > df['lt_ma']
df['Sell'] = df['st_ma'] < df['lt_ma']

df
```

```python
df['Transactions'] = np.where(df['Buy'] != df['Buy'].shift(+1),1,0)
columnsTitles = ['Year', 'Month', 'Date','Day_name','Open','High','Low','Close','Adj Close','Volume','%_change','st_ma','lt_ma','Buy','Sell','Transactions']
df.reindex(columns=columnsTitles)
```

```python
df_moving_average = df[['Year', 'Month', 'Date','Adj Close','Buy','Sell','Transactions']]
df_moving_average
```

```python
def buy(cost,price):
    shares = cost / price
    return shares

cost = 100
    
shares = []

df_moving_average.is_copy = False

for row in range(50,1260):
    if (df_moving_average.loc[row]['Buy'] == True) and (df_moving_average.loc[row]['Transactions'] ==1):
        shares.append(buy(cost,df_moving_average.loc[row]['Adj Close'])) 
    elif (df_moving_average.loc[row]['Buy'] == True) and (df_moving_average.loc[row]['Transactions'] ==0):
        shares.append(shares[len(shares)-1])
    else:
        shares.append(0)



df_moving_average['Shares']= shares

        
df_moving_average
```

```python
def sell(cost,shares, price):
    revenue = shares * price
    profit = revenue - cost
    return profit

profit = [] 

for row in range(50,1260):
    if (df_moving_average.loc[row]['Sell'] == True) and (df_moving_average.loc[row]['Transactions'] == 1):
        profit.append(sell(cost,df_moving_average.loc[row-1]['Shares'],df_moving_average.loc[row]['Adj Close'])) 
    else:
        profit.append(0)


df_moving_average['Profit']= profit
df_moving_average.tail(150)
```

```python
# count the total number of transactions
df_trans = df_moving_average[(df_moving_average['Transactions']==1)]

total_trans = len(df_trans)/2    #1 buy + 1 sell = 1 transaction
total_profit = sum(df_trans['Profit'])
total_average_profit = total_profit / total_trans

print("trans = ", total_trans)
print("P/L = ",total_average_profit )

# count the number of win transactions
df_win_trans = df_trans[df_trans['Profit']> 0]
win_trans = len(df_win_trans)
win_profit = sum(df_win_trans['Profit'])
average_win_profit = win_profit/ win_trans

print("winning = ", win_trans)
print("P/L = ",average_win_profit)

# count the number of lost transactions
df_lose_trans = df_trans[df_trans['Profit'] < 0]
lose_trans = len(df_lose_trans)
lose_profit = sum(df_lose_trans['Profit'])
average_lose_profit = lose_profit/ lose_trans

print("losing = ", lose_trans)
print("P/L = ", average_lose_profit )
```
