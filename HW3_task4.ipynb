### compute the last digit (cent position) in price and plot the histograms (open price, high price, low price, adj close) for your stock. Are last digits distributed evenly (10% end on digit 1, 10% on digits) or not? What digits are the most frequent? Least frequent? 

```python

open_price = list(df['Open'])
high_price = list(df['High'])
low_price = list(df['Low'])
adj_price = list(df['Adj Close'])
```

```python
open_last_digit = []
high_last_digit = []
low_last_digit = []
adj_last_digit = []

for number in open_price:
    number = str(number)
    open_last_digit.append(number[-1])
    
for number in high_price:
    number = str(number)
    high_last_digit.append(number[-1])
    
for number in low_price:
    number = str(number)
    low_last_digit.append(number[-1])
    
for number in adj_price:
    number = str(number)
    adj_last_digit.append(number[-1])
    

```

### Open Price : Not distributed evenly
### Most frequent: 9
### Least frequent: 0

```python
plt.hist(open_last_digit)
plt.title('open price last digit')
```

### High Price : Not distributed evenly
### Most frequent: 1
### Least frequent: 0

```python
plt.hist(high_last_digit)
plt.title('high price last digit')
```

### Low Price : Not distributed evenly
### Most frequent: 8
### Least frequent: 0

```python
plt.hist(low_last_digit)
plt.title('low price last digit')
```

### Adjusted Price : Not distributed evenly
### Most frequent: 2
### Least frequent: 3

```python
plt.hist(adj_last_digit)
plt.title('adj close price last digit')
```
