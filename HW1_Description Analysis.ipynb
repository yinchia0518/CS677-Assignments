```python
# parsing date column
df['Year'] = pd.DatetimeIndex(df['Date']).year
df['Month'] = pd.DatetimeIndex(df['Date']).month
df['Day_name'] = pd.DatetimeIndex(df['Date']).weekday_name
df['Date'] = pd.DatetimeIndex(df['Date']).day
df.head(5) 
```

```python
# compute average close price by year
df_year_close = df[['Year', 'Close']]
avg_by_year = df_year_close.groupby('Year')['Close'].mean()
avg_by_year
```

```python
# compute average close price by year and month
df_year_month_close = df[['Year', 'Month', 'Close']]
avg_by_year_month = df_year_month_close.groupby(['Year', 'Month'])['Close'].mean()
avg_by_year_month
```

```python
# reset index
avg_by_year_month = avg_by_year_month.reset_index()
avg_by_year_month
```

```python
# rename column name(s)
avg_by_year_month.rename(columns = {'Close': 'mean_month'},inplace = True)
avg_by_year_month.head(5)
```

```python
# percent change
df['%_change'] = df['Adj Close'].pct_change()
# cleaning data: replace "NaN" with "0"
df.fillna(0, inplace = True)
df.head(5)
```

```python

# computing moving average (Window=14 days)
w =14

df['std'] = df['Adj Close'].rolling(w).std()
df['st_ma'] = df['Adj Close'].rolling(w).mean()

k=2
df['Lower_Band']= df['st_ma'] - k*df['std']
df['Upper_Band']= df['st_ma'] + k*df['std']

# remove the first 14 days' data, since each moving average need 14 days to compute, NaN cannot be used
df = df.tail(len(df)-13)

# buy and sell signals
df['Buy'] = df['Adj Close'] > df['Upper_Band']
df['Sell'] = df['Adj Close'] < df['Lower_Band']
df
```

```python
columnsTitles = ['Year', 'Month', 'Date','Day_name','Open','High','Low','Adj Close','st_ma','std','Lower_Band','Upper_Band','Buy','Sell']
df = df[columnsTitles]
conditions = [(df['Buy']== True) & (df['Sell'] == False), (df['Buy']== False) & (df['Sell'] == True),(df['Buy']== False) & (df['Sell'] == False)]
choices = [1,1 ,0]
df['Suppose_Trans'] = np.select(conditions,choices)

df_trans = df[df['Suppose_Trans']==1]
df_trans['Real_Trans'] = np.where(df_trans['Buy'] !=df_trans['Buy'].shift(+1),1,0) 
df_trans = df_trans[['Year', 'Month', 'Date','Day_name','Adj Close','Buy','Sell','Suppose_Trans','Real_Trans']]
df_trans

```

```python
def buy(cost,price):
    shares = cost / price
    return shares

cost = 100
    
shares = []

#df_trans.is_copy = False
df_trans = df_trans.reset_index()

for row in range(125):
    if (df_trans.loc[row]['Buy'] == True) and (df_trans.loc[row]['Real_Trans'] ==1):
        shares.append(buy(cost,df_trans.loc[row]['Adj Close'])) 
    elif (df_trans.loc[row]['Buy'] == True) and (df_trans.loc[row]['Real_Trans'] ==0):
        shares.append(shares[len(shares)-1])
    else:
        shares.append(0)
        
df_trans['Shares']= shares

df_trans
```

```python
def sell(cost,shares, price):
    revenue = shares * price
    profit = revenue - cost
    return profit

profit = [] 

for row in range(125):
    if (df_trans.loc[row]['Sell'] == True) and (df_trans.loc[row]['Real_Trans'] == 1):
        profit.append(sell(cost,df_trans.loc[row-1]['Shares'],df_trans.loc[row]['Adj Close'])) 
    else:
        profit.append(0)


df_trans['Profit']= profit
df_trans
```

```python
# count the total number of transactions

total_trans = sum(df_trans['Real_Trans'])/2    #1 buy + 1 sell = 1 transaction
total_profit = sum(df_trans['Profit'])
total_average_profit = total_profit / total_trans

print("trans = ", total_trans)
print("P/L = ",total_average_profit )

# count the number of win transactions
df_win_trans = df_trans[df_trans['Profit']> 0]
win_trans = len(df_win_trans)
win_profit = sum(df_win_trans['Profit'])
average_win_profit = win_profit/ win_trans

print("winning = ", win_trans)
print("P/L = ",average_win_profit)

# count the number of lost transactions
df_lose_trans = df_trans[df_trans['Profit'] < 0]
lose_trans = len(df_lose_trans)
lose_profit = sum(df_lose_trans['Profit'])
average_lose_profit = lose_profit/ lose_trans

print("losing = ", lose_trans)
print("P/L = ", average_lose_profit )
```
