###  Take the last 1,000 trading days (about 4 years of data) and compute the mean m and standard deviation sigma of  daily returns.

```python
df_1000days = df.tail(1000)
df_1000days.rename(columns={'Adj Close': 'Price'}, inplace = True)
df_1000days = df_1000days[['Date', 'Price']]

df_1000days['daily_returns'] = (df_1000days.Price - df_1000days.Price.shift(1))/df_1000days.Price.shift(1)
df_1000days
```

```python
return_mean = round(df_1000days.loc[:,'daily_returns'].mean(),4)
return_std = round(df_1000days.loc[:,'daily_returns'].std(),4)

print('mean:' ,return_mean )
print('standard deviation:' ,return_std )
```

###  If the distribution of daily returns were normal, 5% of days (50 days) will have the daily returns outside of mean +/- 2 * sigma. Compute the number of daily returns outside of mean +/ 2 standard deviations. 

```python
upper_bound = return_mean + 2 * return_std
lower_bound = return_mean - 2 * return_std

print('upper_bound:' ,upper_bound )
print('lower_bound:' ,lower_bound )

count = 0 
for x in df_1000days['daily_returns']:
    if (x < lower_bound) | (x> upper_bound):
        count +=1
print('# of outliers:', count)
```

### Conclusion: There are more than 50 numbers are outside of mean +/- 2 * sigma. So the daily returns is not a normal distribution.
