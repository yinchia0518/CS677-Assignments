```python
def compute_percent_df(df):
    try:
        new_df = df.copy()
        new_df.rename(columns={'Adj Close': 'Price'}, inplace = True)
        new_df = new_df[['Date', 'Price']]
        new_df['price_change(%)'] = new_df['Price'].pct_change() * 100
        new_df.fillna(0, inplace = True)
        return new_df
    except Exception as e:
        print(e)
        return None 
```

### (a)	take your stock (Oct  2, 2013 – Aug 31, 2018)
### (b) compute daily percent and define good/bad days according to %change  

```python
df['Date'] = pd.to_datetime(df['Date']) 
from_ts = '2013-10-02'
to_ts = '2018-08-31'
 
mask = (df['Date'] >= from_ts) & (df['Date'] <= to_ts)
df = df.loc[mask]

new_df = compute_percent_df(df)
new_df['good/bad(1/0)'] = np.where(new_df['price_change(%)'] >= 0,1,0)
new_df = new_df.reset_index()
len(new_df)
```

### (c)	Buy if you have seen W consecutive 0, buy at the end of the W-th day and always sell the next day

```python
# the given df should have 'good/bad(1/0)' column

def compute_sell_buy_df(df,w):
    new_df = df.copy()
    count = 0
    list = df['good/bad(1/0)']
    buy_list = [] # index of row to buy
    sell_list = [] # index or row to sell
    buy_sell_list = []

    for i in range(len(list)):
        if list[i] == 0:
            count +=1
            if count == w:
                count = 0 
                buy_list.append(i)
                sell_list.append(i+1)
                i+=2
            else:
                i+=1
        else:
            count = 0
            i+=1
    
        
    for i in range(len(df)):
        if i in buy_list:
            buy_sell_list.append('buy')
        elif i in sell_list:
            buy_sell_list.append('sell')
        else:
            buy_sell_list.append('hold')
            
    
    new_df['buy/sell'] = buy_sell_list
    
    return new_df


```

```python
def buy(cost,price):
    shares = cost / price
    return shares

def sell(cost,shares, price):
    revenue = shares * price
    profit = revenue - cost
    return profit

def profit_calculator(new_df):
    cost = 100    
    shares = []
    profit = []

        
    for row in range(len(new_df)):
        if new_df.loc[row]['buy/sell'] == 'buy':
            shares.append(buy(cost,new_df.loc[row]['Price']))    
        elif new_df.loc[row]['buy/sell'] == 'sell':
            shares.append(0)      
        else:
            if row ==0:
                shares.append(0)
            else:    
                shares.append(shares[len(shares)-1])
    new_df['Shares']= shares
    
    for row in range(len(new_df)):
        if new_df.loc[row]['buy/sell'] == 'sell':
            profit.append(sell(cost,new_df.loc[row-1]['Shares'],new_df.loc[row]['Price']))
        else:
            profit.append(0)
    
    new_df['Profit']= profit
    
    return new_df
```

```python
def compute_wday_df(df,w):
    ww_df = compute_sell_buy_df(df,w)
    new_ww_df = profit_calculator(ww_df)
    return new_ww_df    
```

```python
for w in range(1,6):
    print('w =', w)
    df = compute_wday_df(new_df,w)
    tran_df = df[df['buy/sell']=='sell']
    transactions = len(tran_df)
    profit = sum(df['Profit'])
    print(' transactions:' , transactions)
    print(' profit:', profit)
```
